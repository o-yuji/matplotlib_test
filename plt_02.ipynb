```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
f1 = 3 * np.sin(x)
f2 = np.sin(2*x)
f3 = 0.3 * np.sin(x)
f4 = np.cos(x)
```

```python
plt.plot(x,f1)
plt.plot(x,f1,'ob')
plt.plot(x,f2)
plt.plot(x,f2,'xr')
plt.plot(x,f3)
plt.plot(x,f3,'+y')
plt.plot(x,f4,color='orchid',linewidth=2,linestyle='--')
plt.plot(x,f4,color='indigo',marker='*')

startx,endx = [0,6.29]
starty,endy = [-3.1,3.1]

plt.axis([startx,endx,starty,endy])

plt.show()
```
