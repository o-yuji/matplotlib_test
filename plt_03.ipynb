```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,25,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.axis([0,25,-2,2])
plt.legend(loc='upper left')

plt.title('sin & cos Plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.show()
```

```python
def sq(x):
    return x**2

def cube(x):
    return x**3

def power4(x):
    return x**4
```

```python
plt.plot(x,sq(x),label='squere')
plt.plot(x,cube(x),label='cube')
plt.plot(x,power4(x),label='power4')
plt.axis([0,25,0,400000])

plt.title('functions')
plt.legend()
plt.ylabel('calc_result')
plt.xlabel('x')
plt.grid(True)

plt.show()
```
